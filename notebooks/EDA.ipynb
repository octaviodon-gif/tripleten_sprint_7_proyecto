### Findings

- Columns   
    - model_year 
        - fill with median by model - 3619
        - type float64 --> int64
    - cylinders
        - null values -5260 into "unknown"
    - odometer
        - fill nulls with median by model+year
        - type float64 --> int64
    - paint_color
        - fill nulls with "unknown"
    - is_4wd
        - Only has 1's and NaN, nulls will be replaced with 0 to properly analyze.

```python
#replacing missing values in 'model_year' with the median
median_model_year = df_vehicles_us['model_year'].median()
#df_vehicles_us['model_year'].fillna(median_model_year, inplace=True) #old version
df_vehicles_us.fillna({'model_year' : median_model_year},inplace=True)

#changing model_year to integer
df_vehicles_us['model_year'] = df_vehicles_us['model_year'].astype('int64')

#replacing missing values in "cylinders" with "unknown"
df_vehicles_us.fillna({'cylinders': 'unknown'}, inplace=True)
#converting odometer to integer with NA handling
df_vehicles_us['odometer'] = df_vehicles_us['odometer'].astype('Int64') 
#replacing odometer missing values with the median
median_odometer = df_vehicles_us['odometer'].median()
df_vehicles_us.fillna({'odometer': median_odometer}, inplace=True)
#changing odometer to integer
df_vehicles_us['odometer'] = df_vehicles_us['odometer'].astype('int64') 

#replacing paint_color missing values with "unknown"
df_vehicles_us.fillna({'paint_color': 'unknown'}, inplace=True)
#replacing is_4wd missing values with 0
df_vehicles_us.fillna({'is_4wd': 0}, inplace=True) 
#changing is_4wd to integer
df_vehicles_us['is_4wd'] = df_vehicles_us['is_4wd'].astype('int64')
```
